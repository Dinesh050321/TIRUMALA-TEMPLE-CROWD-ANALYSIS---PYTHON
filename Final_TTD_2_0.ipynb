```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("/content/TTD data P1.csv")

df.head()
```

```python
df.columns
```

```python
df.rename(columns={
    'date':'Date',
    'total darshans':'Total_Darshans'
}, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
```

```python
df['Is_Weekend'] = df['Weekday'].isin(['Saturday', 'Sunday'])
```

```python
yearly = df.groupby('Year')['darshans'].sum()

plt.figure()
yearly.plot(marker='o')
plt.title("Yearly Darshan Comparison (2013-2025)")
plt.xlabel("Year")
plt.ylabel("Total Darshans")
plt.show()
```

```python
monthly = df.groupby(['Year','Month'])['darshans'].sum().unstack()

monthly.T.plot(kind='area')
plt.title("Monthly Crowd Trend")
plt.xlabel("Month")
plt.ylabel("Darshans")
plt.show()
```

```python
labels = ['Festival Crowd', 'Normal Days']
sizes = [
    df[df['is_festival']==1]['darshans'].sum(),
    df[df['is_festival']==0]['darshans'].sum()
]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.70)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Festival vs Normal Crowd")
plt.show()
```

```python
plt.figure()
plt.scatter(df['rainfall'], df['darshans'])
plt.title("rainfall Impact on Crowd")
plt.xlabel("rainfall (mm)")
plt.ylabel("Total Darshans")
plt.show()
```

```python
plt.figure()
plt.scatter(df['temp_max'], df['darshans'])
plt.title("Temperature Impact on Crowd")
plt.xlabel("Temperature (°C)")
plt.ylabel("Total Darshans")
plt.show()
```
